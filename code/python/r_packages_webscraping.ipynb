```python
from pipelines import pipeline
context_qg = pipeline("multitask-qa-qg", model="valhalla/t5-base-qa-qg-hl")
```

```python
import uuid
import re


error_count = 0
error_context = []
ds_write_count = 0
ds_write_context = []
r_packages_qas_dataset = []
context_dict = {}
qas_list = []
qa_dict = {}
answer_list = []
answer_dict = {}
drop_chars = {"'":None, '"': None, '\n' : ' '}

for item in r_packages_corpus[:20]:

  context_temp = item['package_details'].translate(str.maketrans(drop_chars))
  context = " ".join(context_temp.split())
  
  context_dict['context'] = context
  
  try:
    context_qg_list = context_qg(context)

    # print("-------------------------------")
    # print(context_qg_list)

    for context_qg_qas in context_qg_list:

      # print(context_qg_qas)

      question = context_qg_qas['question']
      answer = context_qg_qas['answer']
      answer_start = re.search(r'\b({})\b'.format(answer.split(' ', 1)[0]), context)

      qa_dict['id'] = str(uuid.uuid4())
      qa_dict['is_impossible'] = False
      qa_dict['question'] = question

      answer_dict['text'] = answer
      answer_dict['answer_start'] = answer_start.start()

      answer_list.append(answer_dict)
      answer_dict = {}

      qa_dict['answers'] = answer_list

      qas_list.append(qa_dict)
      answer_list = []
      qa_dict = {}

    context_dict['qas'] = qas_list
    qas_list = []
    r_packages_qas_dataset.append(context_dict)
    context_dict = {}

    ds_write_count += 1
    # ds_write_context.append(context)

  except:
    error_count += 1
    # error_context.append(context)

    # print(f"!!!!!!    An exception occurred: -->  {context}  !!!!!!!!!")

print(f"qas dataset written count: {ds_write_count}")
print(f"length of qas dataset written: {len(ds_write_context)}")
print(f"qas dataset error count: {error_count}")
```
